## 2.3 PREDICTOR VARIABLES

The "predictor variables" are the features or factors that are used to make predictions about the target variable. Think of it as the pieces of information we feed into our model. 

They are like the clues or hints that our model uses to figure out the best crop recommendation based on the given conditions. The better our predictor variables, the more accurate and helpful our crop recommendations will be.

In this section, we will consider the distribution of the predictor variables to understand the nature of the dataset we are training on.

```python
fig, ax = plt.subplots(3,3, figsize = (12, 10))

fig.subplots_adjust(hspace = 0.5, wspace=0.2)

plt.suptitle("DISTRIBUTION OF PREDICTORS IN THE DATASET", size = 16, y = 0.94);

sns.distplot(x = df ["Nitrogen"], ax = ax[0][0]).set(title = "Nitrogen", xlabel = "Amount")
sns.distplot(x = df ["phosphorus"], ax = ax[0][1]).set(title = "phosphorus", xlabel = "Amount")
sns.distplot(x = df ["potassium"], ax = ax[0][2]).set(title = "potassium", xlabel = "Amount")
sns.distplot(x = df ["temperature"], ax = ax[1][0]).set(title = "temperature", xlabel = "Degree")
sns.distplot(x = df ["ph"], ax = ax[1][1]).set(title = "ph", xlabel = "Amount")
sns.distplot(x = df ["rainfall"], ax = ax[1][2]).set(title = "rainfall", xlabel = "Amount")
sns.distplot(x = df ["humidity"], ax = ax[2][0]).set(title = "humidity", xlabel = "Amount")
```

The insights gained from analyzing the distribution of the predictor variables (Nitrogen, Phosphorus, Potassium, Temperature, Ph, Rainfall, and Humidity) are as follows: 

Nitrogen:
Insight: The distribution of Nitrogen concentrations reveals a predominant range between 20 and 40 ppm, suggesting that most regions exhibit optimal Nitrogen levels for crop growth. 

Phosphorus:
Insight: Phosphorus levels show a diverse distribution, with a concentration peak around 50 and 90 ppm. Crops in regions with higher Phosphorus concentrations demonstrate robust growth, while areas with lower concentrations may benefit from tailored fertilization strategies to improve Phosphorus availability.

Potassium:
Insight: The distribution of Potassium concentrations highlights a range between 30 and 50 ppm, indicating favorable conditions for crops. However, outliers with concentrations exceeding 50 ppm may suggest the need for specific recommendations, such as selecting crops that thrive in Potassium-rich soils or adjusting fertilizer application.

Temperature:
Insight: The distribution of temperatures reflects seasonal variations, with peaks during warmer months. Crops suited to higher temperatures, such as tropical varieties, may be recommended in regions with consistently elevated temperatures, while those adapted to cooler climates could be suggested for cooler months.

PH:
Insight: The distribution of soil pH values shows a predominant range between 20oc and 30oc, indicating neutral to slightly acidic conditions. Crops preferring these pH levels are likely to thrive, but attention should be given to outliers with pH values outside this range to tailor recommendations for specific crops.

Rainfall:
Insight: Rainfall distribution reveals seasonal patterns, with peaks during monsoon months. Regions experiencing higher rainfall may benefit from recommendations for water-loving crops, while those with lower rainfall might be advised to choose drought-resistant varieties.

Humidity:
Insight: Humidity levels exhibit variations throughout the year, with peaks during the summer. Crops resilient to higher humidity conditions may be recommended in these regions, while those adapted to drier conditions could be suggested for areas with lower humidity.

## 4.1 DECISION TREE CLASSIFIER

Imagine making decisions like a smart detective. That's what my project does with something called a Decision Tree. It's like a super-smart flowchart that helps us choose the best options.

```python
model1 = DecisionTreeClassifier()
```

```python
model1.fit(trainX, trainy)
```

```python
pred1 = model1.predict(testX)
pred1
```

```python
performance_table = {}
```

```python
report1 = classification_report(testy, pred1, target_names = encoder.classes_ )

recall = recall_score(pred1, testy, average='micro')
precision = precision_score(pred1, testy, average='micro')
fi_score = f1_score(pred1, testy, average='micro')

performance_table.update({'Decision Tree': [recall, precision, fi_score]})
```

```python
print(performance_table)
```

```python
print(report1)
```
